```python
playgrounds = pd.read_csv("https://data.wprdc.org/datastore/dump/47350364-44a8-4d15-b6e0-5f79ddff9367")  
playgrounds = playgrounds.groupby("neighborhood").count().filter(["id"])
playgrounds = playgrounds.rename(columns={"id":"num_playgrounds"})
playgrounds.sort_values(by=["num_playgrounds"], ascending=False)
```
